## Let's Start with Data Visualization

```python
# Count the number of male and female employees
gender_count = hr_df['Sex'].value_counts()

# Display the count
print(gender_count)

gender_count.plot(kind='bar', color=['Pink', 'Blue'])
plt.xlabel('Gender')
plt.ylabel('Count')
plt.title('Count of Male and Female Employees')
plt.xticks(rotation=0)
plt.show()
```

* The graph shows a higher count of female employees compared to males.
* In summary, with 176 females and 135 males, the gender distribution leans slightly towards females. This distribution can be considered relatively balanced depending on the context and industry norms, but it also highlights opportunities for ensuring equality and inclusivity in organizational practices.


```python
sns.barplot(x = "Department" , y = "Salary", hue = "Department" , data = hr_df)
plt.title('Comparision of Salary as per Department', fontsize=15, fontweight='bold', fontstyle='italic')
sns.set(rc={'figure.figsize':(12,5)})
plt.show
```

* As we can see from the above graph the maximum salary is given to Excutive Office Department.
* The minimum salary is observed in departments like Admin Offices, Sales, and Production.
* This indicates that executive and possibly technical departments are better compensated compared to administrative and production roles.

```python
sns.barplot(x = "Salary" , y = "Position", data = hr_df)
plt.title('Comparision of Salary as per Position', fontsize=12, fontweight='bold', fontstyle='italic')
sns.set(rc={'figure.figsize':(15,15)})
plt.show
```

* Positions such as President & CEO, CIO, Director of sales & IT, tend to have higher median salaries and wider salary ranges.
* Operational and administrative positions, such as Admin Staff, Sales Representative, and Production Worker, tend to have lower median salaries with a narrower range.

```python
# Get the top 10 employees with the highest salary
top_10_employees = hr_df.nlargest(10, 'Salary')[['Employee_Name', 'Salary']]
print(top_10_employees)
```

```python
# Plotting
plt.figure(figsize=(10, 6))
plt.barh(top_10_employees['Employee_Name'], top_10_employees['Salary'], color='skyblue')
plt.xlabel('Salary')
plt.ylabel('Employee Name')
plt.title('Top 10 Employees with Highest Salary',fontsize=12, fontweight='bold', fontstyle='italic')
plt.gca().invert_yaxis()  # To display the highest salary at the top
plt.show()
```

* Among the top 10 highest-paid employees, King Janet and Zamora Jennifer have the highest salaries.


```python
plt.figure(figsize=(12, 8))
sns.barplot(x='Position', y='Salary', hue='Employee_Name', data=top_10_employees, dodge=False)
plt.title('Top 10 Employees Salary Comparison by Position', fontsize=12, fontweight='bold', fontstyle='italic')
plt.xlabel('Position')
plt.ylabel('Salary')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.xticks(rotation=45)
plt.show()
```

* The highest salary is awarded to employees in executive positions, such as CEO and CIO.
* There might be a significant gap between the salaries of the highest-paid employees and the rest, indicating a large difference in compensation at the top levels.
* Comparing the salaries of these top employees with their positions can highlight the roles that are considered most valuable within the company.

```python
# Extract year from DateofHire
hr_df['YearOfHire'] = hr_df['DateofHire'].dt.year

# Count number of hires per year
hires_per_year = hr_df['YearOfHire'].value_counts().sort_index()

# Plotting
plt.figure(figsize=(12, 8))
hires_per_year.plot(kind='line')
plt.title('Hiring Trend Over Years',fontsize=12, fontweight='bold', fontstyle='italic')
plt.xlabel('Year')
plt.ylabel('Number of Hires')
plt.grid(True)
plt.show()
```

* Peaks in the graph from 2010 to 2011 indicate periods of significant hiring, suggesting company expansion or seasonal hiring trends.
* Declines in hiring might correlate with economic downturns, hiring freezes, or periods of company restructuring.
* Consistent hiring trends suggest stable growth, while erratic patterns might need further investigation.

```python
# Plotting
plt.figure(figsize=(12, 8))
sns.scatterplot(x='PerfScoreID', y='DaysLateLast30', data=hr_df)
plt.title('Late Days in Last 30 Days vs Performance Score',fontsize=12, fontweight='bold', fontstyle='italic')
plt.xlabel('Performance Score')
plt.ylabel('Days Late in Last 30 Days')
plt.show()
```

* A negative correlation might be observed where higher performance scores are associated with fewer late days, indicating that punctuality could be a factor in performance evaluation.
* If there is no clear correlation, it suggests that late days might not significantly impact performance scores.
* This analysis can help in understanding if punctuality should be emphasized more in performance evaluations.

```python
# Calculate average engagement survey score by department
avg_engagement_by_dept = hr_df.groupby('Department')['EngagementSurvey'].mean().reset_index()

# Plotting
plt.figure(figsize=(9, 6))
sns.barplot(x='Department', y='EngagementSurvey', data=avg_engagement_by_dept)
plt.title('Average Engagement Survey Score by Department',fontsize=12, fontweight='bold', fontstyle='italic')
plt.xlabel('Department')
plt.ylabel('Average Engagement Survey Score')
plt.xticks(rotation=45)
plt.show()
```

* The engagement scores are higher in the Executive Office and Admin Offices departments compared to others.
* Lower scores in departments like Sales, Production and IT\IS might indicate areas that need better engagement initiatives.
* Understanding these scores can guide HR in developing targeted programs to improve employee engagement across all departments.

```python
# Plotting
plt.figure(figsize=(22, 10))
sns.countplot(x='Position', hue='Sex', data=hr_df)
plt.title('Gender Distribution by Position',fontsize=15, fontweight='bold', fontstyle='italic')
plt.xlabel('Position')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

* Certain positions like area saalees manager, production technician shows a gender imbalance, highlighting areas where diversity efforts could be focused.
* If high-level positions are predominantly held by one gender, it might indicate a need for more inclusive promotion and recruitment practices.
* This analysis helps in assessing gender diversity and identifying roles where gender diversity needs to be improved.

```python
# Calculate average employee satisfaction by marital status
avg_satisfaction_by_marital_status = hr_df.groupby('MaritalDesc')['EmpSatisfaction'].mean().reset_index()

# Plotting
plt.figure(figsize=(12, 8))
sns.barplot(x='MaritalDesc', y='EmpSatisfaction', data=avg_satisfaction_by_marital_status)
plt.title('Average Employee Satisfaction by Marital Status',fontsize=12, fontweight='bold', fontstyle='italic')
plt.xlabel('Marital Status')
plt.ylabel('Average Employee Satisfaction')
plt.xticks(rotation=45)
plt.show()
```

* Variations in satisfaction scores across marital statuses can reveal how personal circumstances might influence job satisfaction.
* Higher satisfaction scores for certain marital statuses might reflect the effectiveness of existing support programs, while lower scores could indicate areas needing more support.
* This insight helps in tailoring employee support programs to better meet the needs of employees based on their marital status.

```python
# Plotting
plt.figure(figsize=(12, 8))
sns.countplot(x='Department', hue='TermReason', data=hr_df)
plt.title('Termination Reasons by Department',fontsize=12, fontweight='bold', fontstyle='italic')
plt.xlabel('Department')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

* Identifying common termination reasons in specific departments can help in addressing underlying issues, such as poor management practices or job dissatisfaction.
* If certain termination reasons are prevalent across multiple departments, it might indicate broader organizational challenges that need to be addressed.
* Understanding these patterns can guide interventions to reduce turnover and improve employee retention.

```python
# Plotting
plt.figure(figsize=(12, 8))
sns.boxplot(x='Sex', y='Absences', data=hr_df)
plt.title('Absences by Gender',fontsize=12, fontweight='bold', fontstyle='italic')
plt.xlabel('Gender')
plt.ylabel('Absences')
plt.show()
```

* If one gender shows significantly more absences, it could point to work-life balance issues, workplace environment challenges, or health-related factors.
* Analyzing these patterns can help the company in implementing policies that address the specific needs of male and female employees, promoting a healthier work environment.
